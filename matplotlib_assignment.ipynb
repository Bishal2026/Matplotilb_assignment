***Q2***


A scatter plot is a type of plot that displays the relationship between two numerical variables. It represents each data point as a marker positioned on a 2D plane.



```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

***Q3***

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.Using the subplot() function is beneficial when you want to display multiple plots or visualizations in a single figure, making it easier to compare and analyze different aspects of the data. 

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.show()
```

***Q4***

A bar plot, also known as a bar chart, is a type of plot that represents categorical data using rectangular bars.

It use to display comparing, visualizing,showing rankings,etc

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit ')
plt.title('Company Profits')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Company Profits')
plt.show()
```

***Q5***

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's summary statistics and its distribution. 


It used for Summary Statistics,Outlier Detection,Comparisons,etc


```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot( [box1, box2])
plt.title('Box Plot')
plt.show()
```
